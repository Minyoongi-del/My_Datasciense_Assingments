Loading the Dataset:

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Cardiotocographic.csv')
df
```

Handling Missing Values:

```python
# Identify Missing Values
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
# Treat Missing Values
df.fillna(df.mean(), inplace=True)  # Simple imputation
# or
df.dropna(inplace=True)  # Remove rows with missing values
print(df.isnull().sum())
```

```python
# Data Type Correction'
df['LB'] = df['LB'].astype(float)
print(df.dtypes)
```

```python
# Detecting and Treating Outliers

## Use statistical methods like the IQR method
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
outliers = df[((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]  # Remove outliers
print("Outliers:")
print(outliers)
```

Statistical Summary

```python
# Summary Statistics
summary = df.describe()
print(summary)

```

Data Visualization

```python
# Histograms for continuous variables
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
df['LB'].hist(bins=30)
plt.title('Distribution of Baseline Fetal Heart Rate (LB)')
plt.xlabel('Baseline Fetal Heart Rate (LB)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plots to identify outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['LB'])
plt.title('Box Plot of Fetal Heart Rate')
plt.show()
```

```python
# Correlation Analysis

## Heatmap to visualize correlations
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pair Plots for Relationships
sns.pairplot(df)
plt.show()
```

Pattern Recognition and Insights

```python
# Identify Correlations
correlation_matrix = df.corr()
print("Correlation Matrix:")
print(correlation_matrix)
```
